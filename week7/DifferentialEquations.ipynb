```python
g=9.8
l=0.5
th0 = 20.*pi/180
t = arange(0.,10.+0.01,0.01)
thetaapp = lambda t: th0*cos(sqrt(g/1)*t)
plot(t,
thetaapp(t))
```

```python
def deriv(thetanow,t):
    g = 9.8
    l = 0.5
    theta = thetanow[0]
    omega = thetanow[1]
    dtheta=omega
    domega=(-g/l)*sin(theta)
    return array([dtheta,domega])

```

```python
from scipy.integrate import odeint

t = arange(0.,10.+0.01,0.01)

omega0 = 0.
thetai = array([th0,omega0])

theta = odeint(deriv,thetai,t)
```

```python
plot(t,thetaapp(t),'b-',t,theta[:,0],'r--')
```

```python
th1 = 0.*pi/180
th2 = 45.*pi/180
th3 = 90.*pi/180
th4 = 135.*pi/180
th5 = 180.*pi/180

thetai1 = array([th1,omega0])
thetai2 = array([th2,omega0])
thetai3 = array([th3,omega0])
thetai4 = array([th4,omega0])
thetai5 = array([th5,omega0])

theta1 = odeint(deriv,thetai1,t)
theta2 = odeint(deriv,thetai2,t)
theta3 = odeint(deriv,thetai3,t)
theta4 = odeint(deriv,thetai4,t)
theta5 = odeint(deriv,thetai5,t)

plot(t,thetaapp(t),color="blue",linewidth = 3)
plot(t,theta1[:,0],color="red",linestyle="--")
plot(t,theta2[:,0],color="orange",linestyle="--")
plot(t,theta3[:,0],color="green",linestyle="--")
plot(t,theta4[:,0],color="cyan",linestyle="--")
plot(t,theta5[:,0],color="purple",linestyle="--")
```

```python
#As the period increases from pi/4 to 3pi/4, the period stretches out so that he two graphs shift out of phase.

#At angle 0 and pi, the graph becomes flat.
```

```python
theta0 = 0.95*pi
t = arange(0.,10.+0.01,0.01)

theta = theta0*cos(sqrt(g/l)*t)
omega = -theta0*(sqrt(g/l))*sin(sqrt(g/l)*t)

plot(t,theta,color="blue")
plot(t,omega,color="red")
```

```python
#Here, the graphs are in phase, and the periods are the same. Omega just starts at a different initial point.
```
