```python
f = lambda x: x**(0.5) + sin(x)
    
x = arange(0.,15.,0.001)

y = f(x)

p = plot(x,y,lw = 3)
ylabel(r"$x^{1/2}+\sin(x)$",fontsize=20)
xlabel(r"$x$",fontsize=20)
xlim(0.,15.)

xmin = 1.7
xmax = 12.4

subx = x[logical_and((x>xmin),(x<xmax))]
ymax = max(f(subx))*1.05
ymin= 0

plot([xmin,xmin],[ymin,ymax])
plot([xmax,xmax],[ymin,ymax])
plot([xmin,xmax],[ymin,ymin])
plot([xmin,xmax],[ymax,ymax])

N = 100000

samplex = xmin + (xmax-xmin)*random_sample(N);
sampley = ymin + (ymax-ymin)*random_sample(N);

plot(samplex[1:200],sampley[1:200],".")

```

```python
mask = (sampley<f(samplex))

sum(mask)

I = (xmax-xmin)*(ymax-ymin)*sum(mask)/mask.size
print I
```

```python
Iexact = 26.5172145362

error = (Iexact - I)/Iexact

print str(error*100) + "%"
```
