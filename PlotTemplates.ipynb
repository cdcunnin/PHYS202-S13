Line Plots
----

```python
y = arange(-5,5,0.1) #set the range of y

E = np.cosh(y) #Equation of E
p = np.sinh(y) #Equation of |p|
```

```python



plot(y,E,color = "red",linewidth = 2,linestyle = "--",label="E")
plot(y,p,color = "red",linewidth = 2,linestyle = "-.",label="|p|")

xlabel("x")
ylabel("y")
title("Potential Energy and Wave Function vs. x-Position",fontsize=20)
grid(True)
legend(loc = "lower left") 

ax = gca()
ax.spines['right'].set_color('none')
```
