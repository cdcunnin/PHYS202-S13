```python
#B)

t = arange(-4,4+0.1,0.1)
f = (1+t)/(t-1)
g = e**(-t)

plot(t,f,color="blue",linewidth=1,label="(1+t)/(t-1)")
plot(t,g,color="red",linewidth=1,linestyle=":",label="exp(-t)")
xlabel("time (seconds)")
ylabel("f(t),g(t)")
xlim(-4,4,1)
ylim(-10,10,1)
xticks(arange(-4,4+0.1,1),('-4','-3','-2','-1','0','1','2','3','4'))
yticks(arange(-10,10+0.1,5),('-10','-5','0','5','10'))
grid(True)
legend(loc = "lower left") 
```

```python
#B) 

def is_anagram(s1,s2):
    if len(s1) == len(s2):
        p1 = array(s1)
        p2 = array(s2)
        w1 = sort(p1, axis=1)
        w2 = sort(p2, axis=1)
        if w1 == w2:
            return True
        else:
            return False
        #If the strings are the same length and their components in alphabetical order look the same, then they are anagrams.
    else:
        return False
    
print is_anagram("gallifrey","freyigall")
```
