$V(x,y) = \frac{kq}{\sqrt{x^2 + (y - d)^2}} - \frac{kq}{\sqrt{x^2 + (y + d)^2}}$

```python
import numpy as np


dx = 0.05
dy = 0.05
x,y = meshgrid(arange(-2,2,dx),arange(-2,2,dy))
               
print x.ndim
print y.shape

k = 1
q = 1
d = 1

V = ((k*q)/((x**2) + (y - d)**2)**1/2) - ((k*q)/((x**2) + (y + d)**2)**1/2)

print V

```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = figure()
ax = fig.gca(projection='3d')
p = ax.plot_surface(x,y,V,rstride=1,cstride=1,linewidth=0,cmap=cm.spring) 
cb = fig.colorbar(p,shrink=0.5) #add a color bar to the plot p
cset = ax.contour(x,y,V,zdir='z',offset=-40,cmap=cm.spring)
cset = ax.contour(x,y,V,zdir='x',offset=-2,cmap=cm.spring)



#rstride is array row scaling
#cstride is array column scaling
#cmap is the colormap
```

```python
levels = array([8,4,2,1,0.5,0.25,0,-0.25,-0.5,-1,-2,-4,-8])
contour(x,y,V,levels,linewidths=4,cmap=cm.spring) #linewidthS. Don't forget the S.
show()
```

Ex 1. Make a 3D plot of the function $G(x,y) = {e^{-x^2 - y^2}}$          over the range -6 < (x,y) < 6 using 51 bins in x and 51 bins in y.

$G(x,y) = e^{-x^2 - y^2} $

```python
dx = 0.2
dy = 0.2

x,y = meshgrid(arange(-6,6,dx),arange(-6,6,dy))

G = e**(-x**2 - y**2)

fig = figure()
ax = fig.gca(projection='3d')
p = ax.plot_surface(x,y,G,rstride=1,cstride=1,linewidth=0,cmap=cm.spectral)
cb = fig.colorbar(p,shrink=0.5)
cset = ax.contour(x,y,G,array([0,0.2,0.4,0.6,0.8,1]),zdir='x',offset=-40,cmap=cm.spectral)
```
