```python
x,y = meshgrid(arange(-3,3,0.037),arange(-3,3,0.037))

T = pointPotential(x,y,1e-9,0,0)

fig = figure()
ax = fig.gca(projection='3d')
p = ax.plot_surface(x,y,T,rstride=1,cstride=1,linewidth=0,cmap=cm.coolwarm) 
cb = fig.colorbar(p,shrink=0.5) 
cset = ax.contour(x,y,T,zdir='z',offset=-40,cmap=cm.coolwarm)
cset = ax.contour(x,y,T,zdir='x',offset=-2,cmap=cm.coolwarm)


```

```python
levels = array([3,2,1,0.5,0.25,0,-0.25,-0.5,-1,-2,-3])
contour(x,y,T,levels,linewidths=4,cmap=cm.coolwarm)
show()
```
