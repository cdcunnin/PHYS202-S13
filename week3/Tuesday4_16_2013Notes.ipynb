```python
#Create a simple plot.

t = arange(0.0, 1.0+0.01, 0.01)
c = np.cos(2*2*pi*t)
s = np.sin(2*2*pi*t)
```

```python
fig = plot(t,c)
plot(t,c) #(x,y)
xlabel("time (s)")
ylabel("voltage (mV)")
title("A simple plot",fontsize=20)
grid(True)
savefig("simple_plot.pdf")
savefig("simple_plot.png")
```

```python
plot(t,s,color = "blue",linewidth = 2,linestyle = "--",label="$cos(4{\pi}t)$")
plot(t,c,color = "red",linewidth = 2,linestyle = ":",label="$sin(4{\pi}t)$") #(x,y)
xlabel("time (s)")
ylabel("voltage (mV)")
title("A simple plot",fontsize=20)
grid(True)
legend(loc = "lower left") #You have to have this to get the plot labels.
#Empty parentheses on legend defaults it to upper right.
savefig("simple_plot2.pdf")
savefig("simple_plot2.png")
```
