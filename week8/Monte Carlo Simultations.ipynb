```python
x=arange(0,pi+0.01,0.01)
f = lambda x: sin (x)/2.
F = lambda x: (cos(0)-cos(x))/2

plot(x,f(x),label = "True distribution",linewidth=3)
plot(x, F(x),label = "Cumulative function",linewidth=3)

xlim(0,3.5)
ylim(0,1.1)

legend(loc=0)
```

```python
import numpy

N=10000

z = numpy.random.random_sample(N);
Finv = lambda z: arccos(cos(0)-2*z)
xsamples= Finv(z);

hist(xsamples,50,normed=True,label="Sampled distribution")
plot(x, f(x), label="True Disyribution", linewidth=3,color="red")

xlim=(0,pi)
ylim(0,0.7)
legend(loc=0)
```

```python
#A)

x=arange(0.,5.,0.01)
f = lambda x: q * e**(-q*x)
F = lambda x: q - e**(-q*x)
h = np.cumsum(f(x))
array = h/sum(f(x))

z = numpy.random.random_sample(N);

loc = array.searchsorted(z)

xsamples= x[loc]
```

```python
#B)

hist(xsamples,50,normed=True,label="Sampled distribution")
plot(x,array)

```

```python
#C)

def distribute1D(x,prob,N):
    h = np.cumsum(prob)
    array = h/sum(prob)
    z = numpy.random.random_sample(N);

    loc = array.searchsorted(z)

    xsamples = x[loc]
    
    return xsamples

```
