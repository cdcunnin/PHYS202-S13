```python
#Monte Carlo Methods in-class stuff in week8 folder.

def distribute1D(x,prob,N):
    h = np.cumsum(prob)
    array = h/sum(prob)
    z = numpy.random.random_sample(N);

    loc = array.searchsorted(z)

    xsamples = x[loc]
    
    return xsamples




x = arange(-0.1,0.1,0.0001)
d = 15*(10**-6)
a = 10*(10**-6)
L=1.
w = 632.8*(10**-9)
A= 1

I = ((A**2)*(sinc(a*x/(w*L)))**2) * ((A**2)*(cos(2.*pi*d*x/(w*L)))**2)

print distribute1D(x,I,100)

plot(x,I*300*max(distribute1D(x,I,10000)),label='True distribution')
hist(distribute1D(x,I,10000),50,normed=True, label='Sampled Data')
show()
```
